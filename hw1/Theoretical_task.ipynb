# Теоретические задачи

## Задача 1

Задача баесовского классификатора - найти $P(y|x^1....x^N)$, где:
$$P(y|x^1....x^N) = P(y) \prod_{k = 1}^N P(x^k|y)$$
По условию задачи:
$$P(x^{(k)}|y) = \frac{1}{\sqrt{2\pi\sigma^2}} e^{-\frac{(x^{(k)} - \mu_{yk})^2}{2\sigma^2}}$$
Подставим их в $P(y|x^1....x^N)$:

$$P(y|x^1....x^N) =   P(y) \prod_{k = 1}^N \frac{1}{\sqrt{2\pi\sigma^2}} e^{-\frac{(x^{(k)} - \mu_{yk})^2}{2\sigma^2}} = Const  P(y) e^{-\frac{\sum_{k=1}^N(x^{(k)} - \mu_{yk})^2}{2\sigma^2}}$$
$P(y|x^1....x^N)$ максимальна тогда, когда максимальна степень экспоненты, т.е. минимальна $\sum_{k=1}^N(x^{(k)} - \mu_{yk})^2$ - евклидово расстояние до центра класса $\mu_y$.

## Задача 2

Зафиксируем точки на осях: (0,0) и (1,1). Для нахождения треугольного $ROC-AUC$ необходимо найти третью точку.
Найдём её координаты по значениям $TPR$ и $FPR$ для некотрый выборки $n$:

$TPR = \frac{TP}{TP + FN} = \frac{pu}{n} = p$

$FPR =\frac{FP}{FP + TN} =\frac{p(1 - n)}{(1 - n)} = p$

Тогда координаты третьей точки (p,p).
Через точки (0,0), (p,p) и (1,1) мы сможем провести только прямую $y=x$.
Тогда площадь под этой прямой на отрезке $[0,1]$ равняется $\frac{1}{2}$.

## Задача 4

Пусть $N$ - количество элементов в листе; $y$ - истинный таргет, некоторого элемента; $y_i$ - таргет случайного объекта из листа.

Найдём $MSE_{mean}$ и $MSE_{random}$:

$$MSE_{mean} = (y - \frac{1}{N}\sum_{i = 1}^N y_i)^2$$

Для некоторого элемента i, получим следующее выражение для ошибки
$$SE_{rand} = (y - y_{i})^2$$

$$MSE_{rand} = \frac{1}{N} \sum_{i = 1}^N (y - y_{i})^2 = (y - \frac{1}{N}\sum_{i = 1}^N y_i)^2 = MSE_{mean}$$

Таким образом, данные стратегии являются идентичными.

## Задача 5

Почему такая стратегия не даёт ощутимого выигрыша?

Потому что мы уже сильно сузили область поиска, в этот лист попадают элементы с примерным значениям target'a равным среднему в листе. То есть и регрессию мы будем строить только по ним $\rightarrow$ будем получать ответы с target'ом примерно равным среднему в листе.(Т.к. все таргеты примерно одинаковые в листе, то зависимость в линейной регресии от признаков будет очень слабая, то есть с маленькими кожфициентами).

Как модифицировать?

Дерево должно быть достаточно низким, чтобы число признаков по которым производится отсев было меньше общего числа информативных признаков. Линейную регрессию в листе нужно строить именно по оставшимся признакам, не использоваными при построении пути к этому листу.

Данная стратегия не даёт ощутимого выйгрыша, так как при обучении дерева в лист попадает малая часть обучающей выборки с очень схожим таргетом.
Поэтому результатом регрессии, скорее всего, будет среднее значение таргетов доли обучающей выборки, оказавшейся в листе.

В качестве модификации, перед построением линейной модели, можно использовать pruning для увеличения MSE в листах (и как следствие выборки с различными таргетами).
